```python
from scipy.io import wavfile
import numpy as np
import os
import sys


def load_dataset(dataPathRoot):
    
    dataset = []
    for dirName,subdirList, fileList in os.walk(dataPathRoot):
        for fname in fileList:
            fullFilename = os.path.join(dirName, fname)
                
            if fname.endswith(".WAV"):
                try:
                    (fs,data) = wavfile.read(fullFilename)
                except ValueError as e:
                    print('You should use nist2wav.sh to convert NIST format file to WAV file first')
                
    return dataset
                
    
dataset = load_dataset("/home/erika/Documents/Licenta/datasets/data/lisa/data/timit/raw/TIMIT/TRAIN")

print(len(dataset))
```
